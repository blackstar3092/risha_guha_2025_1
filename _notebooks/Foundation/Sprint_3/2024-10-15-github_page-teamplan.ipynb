# üõí Dnero Store Project Overview
This project aims to develop an interactive online chat room where users, particularly students, can buy their favorite items at the Dnero Store. The platform promotes new items at the Dnero Store, and encourages communities with people who are interested in buying the same thing.

---

## üéâ **Key Innovation**
The distinctive aspect of the project is its **community-building approach**. As students choose items they're interested in, they are automatically placed into groups (chat communities) with others who share similar preferences. This feature boosts user engagement by creating spaces where members can share images and discuss newly available items.

---

## üõ†Ô∏è **Key Features**
- **Activity Chatroom**: A designated space where students share opinions on items sold at the **Dnero Store**.
- **Community Matching**: Automatic group chat creation based on item preferences.
